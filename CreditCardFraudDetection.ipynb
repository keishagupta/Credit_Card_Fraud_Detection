```python
# Credit Card Fraud Detection - Comprehensive Analysis
# No installation needed in Colab for these
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, roc_curve, auc, accuracy_score
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
plt.style.use('seaborn-v0_8-whitegrid')

print("# Credit Card Fraud Detection Project")
print("## Loading and Exploring Data")

# Load dataset from OpenML (a Kaggle mirror)
url = "https://www.openml.org/data/get_csv/1673544/creditcard.csv"
df = pd.read_csv(url)

# Fix column with quoted strings
df['Class'] = df['Class'].str.strip("'").astype(int)

# Quick check
print("Dataset shape:", df.shape)
print("First 5 rows:")
display(df.head())

# Data Information
print("\nData Types:")
display(df.dtypes)

print("\nSummary Statistics:")
display(df.describe())

print("\nMissing values:", df.isnull().sum().sum()) #sum() â†’ total per column, another sum() â†’ total across all columns.
print("Class distribution:")
class_dist = df['Class'].value_counts()
display(class_dist)

# Calculate imbalance ratio
imbalance_ratio = class_dist[0] / class_dist[1]
print(f"Imbalance ratio (Normal:Fraud) = {imbalance_ratio:.2f}:1")

# Distribution of transaction amount by class
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)   #left subplot
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, kde=True, color='blue', alpha=0.6)
plt.title('Distribution of Transaction Amounts (Normal)')
plt.xlabel('Amount')
plt.xlim([0, 500])
plt.ylabel('Frequency')

plt.subplot(1, 2, 2) #RIGHT SUBPLOT
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, kde=True, color='red', alpha=0.6)
plt.title('Distribution of Transaction Amounts (Fraud)')
plt.xlabel('Amount')
plt.ylabel('Frequency')

plt.tight_layout() #Ensures the subplots donâ€™t overlap and are laid out nicely.


plt.show()

# Transaction time analysis
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)  #for normal transactions
sns.histplot(df[df['Class'] == 0]['Time'], bins=50, kde=True, color='blue', alpha=0.6)
plt.title('Distribution of Transaction Time (Normal)')
plt.xlabel('Time (seconds from first transaction)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)# for fraud transactions
sns.histplot(df[df['Class'] == 1]['Time'], bins=50, kde=True, color='red', alpha=0.6)
plt.title('Distribution of Transaction Time (Fraud)')
plt.xlabel('Time (seconds from first transaction)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Convert time to hours for better interpretability
df['Hour'] = np.round(df['Time'] / 3600) % 24

# Fraud by hour of day
plt.figure(figsize=(14, 6))
hourly_fraud = df.groupby('Hour')['Class'].mean()
sns.lineplot(x=hourly_fraud.index, y=hourly_fraud.values)
plt.title('Fraud Rate by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Fraud Rate')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()

print("\n## Feature Analysis")

# Explore only selected anonymous features (top correlated with fraud)
plt.figure(figsize=(16, 10))
selected_features = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8']
for i, feature in enumerate(selected_features):
    plt.subplot(2, 4, i+1)
    sns.boxplot(x='Class', y=feature, data=df)
    plt.title(f'Box Plot of {feature} by Class')
plt.tight_layout()
plt.show()

# Correlation matrix
plt.figure(figsize=(16, 14))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, linewidths=0.5)  #Red = high positive correlation, Blue = high negative correlation.
plt.title('Correlation Matrix of All Features')
plt.show()

# Top correlated features with the target variable
corr_with_target = corr_matrix['Class'].sort_values(ascending=False)
print("\nTop 10 features correlated with fraud:")
display(corr_with_target[1:11])  # Skip the first one (Class correlation with itself)

print("\nBottom 10 features correlated with fraud:")
display(corr_with_target[-10:])

print("\n## Data Preprocessing")

# Normalize 'Amount' using RobustScaler (less sensitive to outliers)
robust_scaler = RobustScaler()
df['Amount_scaled'] = robust_scaler.fit_transform(df[['Amount']])

# Standard scaling for all V features
standard_scaler = StandardScaler()
V_cols = [col for col in df.columns if col.startswith('V')]
df[V_cols] = standard_scaler.fit_transform(df[V_cols])

# Drop 'Time' but keep 'Hour' and original 'Amount' for reference
df_processed = df.drop(['Time'], axis=1)

print("Preprocessed data:")
display(df_processed.head())

print("\n## Dimensionality Reduction")

# Apply PCA for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_processed.drop(['Class', 'Hour', 'Amount'], axis=1))
df_processed['pca1'] = pca_result[:, 0]
df_processed['pca2'] = pca_result[:, 1]

# Explained variance
print(f"Explained variance by the first two components: {sum(pca.explained_variance_ratio_):.2f}")

# PCA components explanation
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_processed, x='pca1', y='pca2', hue='Class',
                palette={0: 'blue', 1: 'red'}, alpha=0.6, s=15)
plt.title("PCA Projection of Transactions")
plt.legend(labels=['Normal', 'Fraud'])
plt.show()

# Feature importance based on PCA
components_df = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(2)],
    index=[col for col in df_processed.columns if col not in ['Class', 'Hour', 'Amount', 'pca1', 'pca2']]
)

plt.figure(figsize=(12, 10))
sns.heatmap(components_df, cmap='viridis', annot=False)
plt.title('Feature Importance in PCA Components')
plt.show()

print("\n## Cluster Analysis")

# K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df_processed['kmeans_cluster'] = kmeans.fit_predict(df_processed.drop(['Class', 'Hour', 'Amount', 'pca1', 'pca2'], axis=1))

# Evaluate K-means clustering
print("K-means clustering evaluation:")
kmeans_acc = np.mean(df_processed['kmeans_cluster'] == df_processed['Class'])
print(f"Accuracy (after alignment): {max(kmeans_acc, 1-kmeans_acc):.4f}")

print("\nConfusion Matrix (K-means):")
display(pd.crosstab(df_processed['Class'], df_processed['kmeans_cluster'],
                   rownames=['Actual'], colnames=['Predicted']))

# Visualize clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_processed, x='pca1', y='pca2', hue='kmeans_cluster',
                palette='viridis', alpha=0.6, s=15)
plt.title("Clusters Found by K-Means")
plt.show()
```

```python
print("\n## Anomaly Detection")

# Isolation Forest
iso_forest = IsolationForest(contamination=0.002, random_state=42)  # Set contamination close to actual fraud rate
df_processed['iso_anomaly'] = iso_forest.fit_predict(df_processed.drop(['Class', 'Hour', 'Amount', 'pca1', 'pca2', 'kmeans_cluster'], axis=1))
df_processed['iso_anomaly'] = df_processed['iso_anomaly'].map({1: 0, -1: 1})  # Convert: -1 (anomaly) to 1, 1 (normal) to 0
# Evaluation
print("Isolation Forest Results:")
print("\nConfusion Matrix:")
cm_iso = pd.crosstab(df_processed['Class'], df_processed['iso_anomaly'],
                   rownames=['Actual'], colnames=['Predicted'])
display(cm_iso)

```

```python
print("\nClassification Report:")
print(classification_report(df_processed['Class'], df_processed['iso_anomaly'], zero_division=0))

# Visualize anomalies
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_processed, x='pca1', y='pca2', hue='iso_anomaly',
                palette={0: 'green', 1: 'purple'}, alpha=0.6, s=15) #s is the size of marker in scatterplot
plt.title("Anomaly Detection with Isolation Forest")
plt.legend(labels=['Normal', 'Anomaly'])
plt.show()

print("\n## Supervised Learning")

# Split the data
X = df_processed.drop(['Class', 'Hour', 'Amount', 'pca1', 'pca2', 'kmeans_cluster', 'iso_anomaly'], axis=1)
y = df_processed['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")

# Apply SMOTE to handle class imbalance - but use a smaller sample size for efficiency
# First, reduce the majority class for faster processing
fraud_indices = y_train[y_train == 1].index
non_fraud_indices = y_train[y_train == 0].index[:10000]  # Limit to 10,000 non-fraud examples
balanced_indices = np.concatenate([fraud_indices, non_fraud_indices])

X_train_balanced = X_train.loc[balanced_indices]
y_train_balanced = y_train.loc[balanced_indices]

# Then apply SMOTE on this reduced dataset
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_balanced, y_train_balanced)

print(f"After balancing and SMOTE, training set: {X_train_resampled.shape[0]} samples")
print(f"Class distribution after SMOTE: {pd.Series(y_train_resampled).value_counts().to_dict()}")

# Define models to compare (excluding SVM as it's too slow for this large dataset)
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}

# Train and evaluate models
results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train_resampled, y_train_resampled)

    # Predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred)

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)

    # Precision-Recall Curve
    precision, recall, _ = precision_recall_curve(y_test, y_proba)
    pr_auc = auc(recall, precision)
```

```python
results[name] = {
        'accuracy': accuracy,
        'conf_matrix': conf_matrix,
        'report': report,
        'fpr': fpr,
        'tpr': tpr,
        'roc_auc': roc_auc,
        'precision': precision,
        'recall': recall,
        'pr_auc': pr_auc
    }

print(f"Accuracy: {accuracy:.4f}")
print(f"AUC-ROC: {roc_auc:.4f}")
print(f"AUC-PR: {pr_auc:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(report)

# Plot ROC curves
plt.figure(figsize=(10, 8))
for name, result in results.items():
    plt.plot(result['fpr'], result['tpr'], label=f"{name} (AUC = {result['roc_auc']:.4f})")

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Different Models')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Plot Precision-Recall curves
plt.figure(figsize=(10, 8))
for name, result in results.items():
    plt.plot(result['recall'], result['precision'], label=f"{name} (AUC = {result['pr_auc']:.4f})")

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves for Different Models')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

print("\n## Feature Importance")

# Random Forest feature importance
rf_model = models['Random Forest']
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance.head(15))
plt.title('Top 15 Features by Importance (Random Forest)')
plt.tight_layout()
plt.show()

print("\nTop 15 most important features:")
display(feature_importance.head(15))

print("\n## Model Fine-Tuning")

# Simplified hyperparameter tuning for the Random Forest with fewer options
param_grid = {
    'n_estimators': [100],
    'max_depth': [10, None]
}

grid_search = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=3,
    scoring='roc_auc',
    n_jobs=-1
)

print("Performing simplified grid search for Random Forest...")
grid_search.fit(X_train_resampled, y_train_resampled)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best ROC-AUC score: {grid_search.best_score_:.4f}")

# Final model with tuned parameters
best_rf = grid_search.best_estimator_
y_pred_tuned = best_rf.predict(X_test)
y_proba_tuned = best_rf.predict_proba(X_test)[:, 1]

print("\nFinal tuned model performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_tuned):.4f}")

fpr_tuned, tpr_tuned, _ = roc_curve(y_test, y_proba_tuned)
roc_auc_tuned = auc(fpr_tuned, tpr_tuned)
print(f"ROC-AUC: {roc_auc_tuned:.4f}")

precision_tuned, recall_tuned, _ = precision_recall_curve(y_test, y_proba_tuned)
pr_auc_tuned = auc(recall_tuned, precision_tuned)
print(f"PR-AUC: {pr_auc_tuned:.4f}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_tuned))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_tuned))

print("\n## Cost Analysis")
```

```python
# Function to calculate cost
def calculate_fraud_cost(y_true, y_pred, cost_fp=1, cost_fn=100):
    """
    Calculate the cost of fraud detection
    - cost_fp: cost of false positive (blocking a legitimate transaction)
    - cost_fn: cost of false negative (letting a fraudulent transaction through)
    """
    cm = confusion_matrix(y_true, y_pred)

    # Extract values from confusion matrix
    tn, fp, fn, tp = cm.ravel()

    # Calculate costs
    cost_false_positives = fp * cost_fp
    cost_false_negatives = fn * cost_fn
    total_cost = cost_false_positives + cost_false_negatives

    return {
        'false_positives': fp,
        'false_negatives': fn,
        'cost_false_positives': cost_false_positives,
        'cost_false_negatives': cost_false_negatives,
        'total_cost': total_cost
    }

# Calculate costs for all models
print("Cost analysis (assuming FN costs 100x more than FP):")
for name, model in models.items():
    y_pred = model.predict(X_test)
    cost = calculate_fraud_cost(y_test, y_pred)
    print(f"\n{name}:")
    print(f"False Positives: {cost['false_positives']}")
    print(f"False Negatives: {cost['false_negatives']}")
    print(f"Cost of False Positives: ${cost['cost_false_positives']}")
    print(f"Cost of False Negatives: ${cost['cost_false_negatives']}")
    print(f"Total Cost: ${cost['total_cost']}")

# Cost for tuned model
tuned_cost = calculate_fraud_cost(y_test, y_pred_tuned)
print("\nTuned Random Forest:")
print(f"False Positives: {tuned_cost['false_positives']}")
print(f"False Negatives: {tuned_cost['false_negatives']}")
print(f"Cost of False Positives: ${tuned_cost['cost_false_positives']}")
print(f"Cost of False Negatives: ${tuned_cost['cost_false_negatives']}")
print(f"Total Cost: ${tuned_cost['total_cost']}")

print("\n## Threshold Optimization")

# Find optimal threshold for cost minimization
thresholds = np.linspace(0.01, 0.99, 99)
costs = []

for threshold in thresholds:
    y_pred_threshold = (y_proba_tuned >= threshold).astype(int)
    cost = calculate_fraud_cost(y_test, y_pred_threshold)
    costs.append((threshold, cost['total_cost']))

costs_df = pd.DataFrame(costs, columns=['Threshold', 'Total Cost'])
optimal_threshold = costs_df.loc[costs_df['Total Cost'].idxmin(), 'Threshold']

print(f"Optimal threshold for cost minimization: {optimal_threshold:.4f}")

# Apply optimal threshold
y_pred_optimal = (y_proba_tuned >= optimal_threshold).astype(int)

# Evaluate with optimal threshold
print("\nPerformance with optimal threshold:")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_optimal))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_optimal))

optimal_cost = calculate_fraud_cost(y_test, y_pred_optimal)
print(f"\nOptimal Total Cost: ${optimal_cost['total_cost']}")
```

```python
# Plot cost by threshold
plt.figure(figsize=(10, 6))
plt.plot(costs_df['Threshold'], costs_df['Total Cost'])
plt.axvline(x=optimal_threshold, color='r', linestyle='--',
           label=f'Optimal Threshold = {optimal_threshold:.4f}')
plt.xlabel('Threshold')
plt.ylabel('Total Cost')
plt.title('Cost vs. Classification Threshold')
plt.grid(True)
plt.legend()
plt.show()

print("\n## Summary and Conclusions")

print("1. Dataset Characteristics:")
print(f"   - Total transactions: {len(df)}")
print(f"   - Fraud cases: {class_dist[1]} ({class_dist[1]/len(df)*100:.4f}%)")
print(f"   - Imbalance ratio: {imbalance_ratio:.2f}:1")

print("\n2. Best Model Performance:")
print(f"   - Algorithm: Tuned Random Forest")
print(f"   - Parameters: {grid_search.best_params_}")
print(f"   - Accuracy: {accuracy_score(y_test, y_pred_optimal):.4f}")
print(f"   - ROC-AUC: {roc_auc_tuned:.4f}")
print(f"   - PR-AUC: {pr_auc_tuned:.4f}")

print("\n3. Cost Analysis:")
print(f"   - Optimal threshold: {optimal_threshold:.4f}")
print(f"   - False Positives: {optimal_cost['false_positives']}")
print(f"   - False Negatives: {optimal_cost['false_negatives']}")
print(f"   - Total Cost: ${optimal_cost['total_cost']}")

print("\n4. Most Important Features:")
for i, (feature, importance) in enumerate(zip(feature_importance['Feature'].head(5),
                                            feature_importance['Importance'].head(5))):
    print(f"   {i+1}. {feature}: {importance:.4f}")

print("\n5. Recommendations:")
print("   - Implement the tuned Random Forest model with the optimal threshold")
print("   - Focus on monitoring the top features for fraud detection")
print("   - Consider collecting more data on fraudulent transactions")
print("   - Implement real-time anomaly detection based on the isolation forest model")
```
